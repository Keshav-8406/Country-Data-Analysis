```python
import requests
import pandas as pd
```

# Task 1 — Fetch and Validate the Response

```python
# API endpoint
url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    # Sending GET request to the API
    response = requests.get(url)

    # Printing the HTTP status code to check if request worked
    print("Status Code:", response.status_code)

    # Checking if request is successful (200 = OK)
    if response.status_code == 200:

        # Converting JSON response into Python data (list of countries)
        data = response.json()

        # Printing total number of countries fetched
        print("Total Countries:", len(data))
        print("\nData fetched successfully!")

    else:
        # If request failed, show proper error message with status code
        print(f"Error: Failed to fetch data (Status Code {response.status_code})")

# Handling errors like internet issue, timeout, etc.
except requests.exceptions.RequestException as e:
    print("Exception occurred while making request:", e)
```

# Task 2 — Extract and Display Key Field

```python
# Creating an empty list to store cleaned country data
country_list = []

# Looping through each country received from API
for country in data:

    # Extracting country name (common name like India, USA)
    # Using .get() to avoid errors if key is missing
    name = country.get("name", {}).get("common", "N/A")

    # Extracting population (default = 0 if not available)
    population = country.get("population", 0)

    # Extracting region (Asia, Europe, Africa, etc.)
    region = country.get("region", "N/A")

    # Extracting area in square kilometers (default = 0)
    area = country.get("area", 0)

    # Appending extracted data into list as a dictionary
    country_list.append({
        "common_name": name,
        "population": population,
        "region": region,
        "area": area
    })

# Converting list of dictionaries into a Pandas DataFrame
df = pd.DataFrame(country_list)
df.info()
print("\nDataFrame created successfully!")

# Displaying first 10 rows of the DataFrame
df.head(10)
```

# Task 3 — Summarize the Data

**1. Region with Highest Population**

```python
# Grouping countries by region and summing their populations
region_pop = df.groupby("region")["population"].sum()

# Finding which region has the maximum population
highest_region = region_pop.idxmax()

# Printing the final answer
print(f"Region with Highest Population: {highest_region}")
```

**2. Country with Largest Area**

```python
# Title for output
print("\nTop 10 Countries by Area:\n")

# Showing top countries by area
print(df.sort_values(by="area", ascending=False)[["common_name", "area"]].head(10))

# Finding the country having maximum area
largest_country = df.loc[df["area"].idxmax(), "common_name"]

# Printing the final answer
print("\nCountry with Largest Area:", largest_country)
```
